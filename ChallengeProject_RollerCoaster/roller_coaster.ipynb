```
# Do not have access to this library so commenting it out.
# Creates a few errors, but nothing dramatic, charts still show
#import codecademylib3_seaborn
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# load rankings data here:
wood = pd.read_csv('Golden_Ticket_Award_Winners_Wood.csv')
steel = pd.read_csv('Golden_Ticket_Award_Winners_Steel.csv')

# write function to plot rankings over time for 1 roller coaster here:
def plot_rankings(name, df):

  plt.clf()

  filteredList = df[df.Name == name]

  parks = filteredList.Park.unique()

  for park in parks:
    #print(park)
    rankByPark = filteredList[filteredList.Park == park]
    #print(rankByPark)
    plt.plot(rankByPark["Year of Rank"], rankByPark["Rank"])


  plt.xlabel("Year of Rank")
  plt.ylabel("Ranking")
  plt.title("Historical Ranking of "+name)
  ax = plt.subplot()
  ax.invert_yaxis()

  plt.show()

plot_rankings("El Toro", wood)

# write function to plot rankings over time for 2 roller coasters here:
def plot_multi_rankings(names, df):

  plt.clf()

  #filteredList = df.loc[(df['Name'] == name1) | (df['Name'] == name2)]

  filteredList = df[df['Name'].isin(names)]
  #print(filteredList)

  parks = filteredList.Park.unique()


  legend = []
  for park in parks:
    #print(park)
    rankByPark = filteredList[filteredList.Park == park]
    #print(rankByPark)
    plt.plot(rankByPark["Year of Rank"], rankByPark["Rank"])
    legend.append(park)


  plt.xlabel("Year of Rank")
  plt.ylabel("Ranking")
  plt.title("Historical Ranking of Coasters")
  #plt.legend(legend)
  #plt.legend(legend, loc=(1,0))
  # Can't figure out how to display the legend properly. Not sure if it
  # is an issue with the Codecademy browser window
  plt.legend(legend, bbox_to_anchor=(1.04,0.5))
  ax = plt.subplot()
  ax.invert_yaxis()

  plt.show()


plot_multi_rankings(["El Toro", "Boulder Dash"], wood)


# write function to plot top n rankings over time here:
def plot_top_n(rank, df):

  filteredList = df[df.Rank <= rank]
  coasterNames = filteredList.Name.unique()

  plot_multi_rankings(coasterNames, filteredList)



plot_top_n(5, steel)

# load roller coaster data here:
rollerCoasters = pd.read_csv('roller_coasters.csv')


# write function to plot histogram of column values here:
def plot_coaster_data(column, df):

  plt.clf()

  #print(df[column].head(50))
  data = df[column].dropna().values
  #print(data)

  plt.hist(data)
  plt.xlabel(column)
  plt.ylabel("Number of Coasters")
  plt.title('Histogram of '+column)
  
  plt.show()
  

plot_coaster_data('speed', rollerCoasters)

# write function to plot inversions by coaster at a park here:
def plot_num_inversions(parkName, df):

  plt.clf()

  filteredList = df[df.park == parkName]
  filteredList = filteredList.sort_values('num_inversions', ascending=False)

  plt.bar(range(len(filteredList.name)), filteredList.num_inversions.values)
  plt.xlabel("Number of Inversions")
  plt.ylabel("Number of Coasters")
  plt.title('Histogram of Number of Inversions at '+parkName)
  ax = plt.subplot()
  ax.set_xticks(range(len(filteredList)))
  ax.set_xticklabels(filteredList.name, rotation=45)
  plt.show()
  

plot_num_inversions('Hersheypark', rollerCoasters)
    
# write function to plot pie chart of operating status here:
def plot_status(df):

  plt.clf()

  filteredList = df['status'].unique()
  
  operating = 0
  closedDefinitely = 0
  for index, row in df.iterrows():
    if(row['status'] == 'status.operating'):
      operating += 1
    elif(row['status'] == 'status.closed.definitely'):
      closedDefinitely += 1
  
  labels = ['status.operating', 'status.closed.definitely']
  explode = (0, 0.1)
  plt.pie([operating, closedDefinitely], labels=labels,  explode=explode, autopct='%0.1f%%')
  plt.axis('equal')
  plt.title("Ratio of Roller Coaster Operating Status")
  
  plt.show()


plot_status(rollerCoasters)

  
# write function to create scatter plot of any two numeric columns here:
def plot_two_nums(cols, df):
  
  plt.clf()

  xlabel = cols[0].capitalize()
  ylabel = cols[1].capitalize()
  plt.legend (cols[0], cols[1])
  plt.scatter(df[cols[0]], df[cols[1]], alpha=0.5)
  plt.xlabel(xlabel)
  plt.ylabel(ylabel)
  plt.title('Scatter Plot of '+xlabel+" & "+ylabel)
  plt.show()

plot_two_nums(['height', 'length'], rollerCoasters)
```
